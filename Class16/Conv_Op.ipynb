```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import cv2
```

```python
im = cv2.imread('./img.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(im, cmap='gray')
plt.show()
print im.shape
```

```python
## Selecting a kernel matrix
K = np.array([[0, -2, 0], [-2, 8, -2], [0, -2, 0]])
print K.shape
print K
```

```python
small_matrix = K
large_matrix = im
```

```python
out_r = im.shape[0] - K.shape[0] + 1
out_c = im.shape[1] - K.shape[1] + 1
new_img = np.zeros((out_r, out_c))
print new_img.shape
```

```python
for rx in range(out_r):
    for cx in range(out_c):
        patch = im[rx:rx+K.shape[0], cx:cx+K.shape[1]]
        prod = patch*K
        prod = prod.sum()
        new_img[rx, cx] = prod
```

```python
plt.imshow(new_img*(new_img>0), cmap='gray')
plt.show()
```
