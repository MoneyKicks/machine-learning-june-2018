```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
old_mask_path = './mask_wo_index/anf_64540_01_model1_Mask_front_inspect.png'
new_mask_path = './images/anf_64540_01_model1_Mask_front_inspect.png'
```

```python
m1 = Image.open(old_mask_path)
plt.imshow(np.array(m1))
plt.show()
print np.unique(np.array(m1), return_counts=True)
```

```python
m2 = Image.open(new_mask_path)
plt.imshow(np.array(m2))
plt.show()
```

```python
old_m1 = np.array(m1)
dil = cv2.dilate(old_m1, np.ones((5, 5)), iterations=1)
plt.imshow(dil)
plt.show()
ero = cv2.erode(old_m1, np.ones((5,5)), iterations=1)
```

```python
bndr = dil - ero
plt.imshow(bndr)
plt.show()
```

```python
print np.unique(bndr)
```

```python
new_bndr = np.array(bndr!=0)*7
plt.imshow(new_bndr)
```
